## tl;dr

Run the notebook top-to-bottom to calculate the provisional year-one ROI scenarios for the Executive Council Steward. The model is assumption-driven because the pilot has not established a measured baseline.

## Context & Methods

Decision: whether the expected value justifies activating and measuring a 30-day Council Steward pilot.

### Key Assumptions

- Six current portfolio projects and 458 repository project files are observed complexity indicators, not direct value multipliers.
- Benefit includes reconstruction time saved and routine rework avoided.
- Cost includes human review time, AI runtime/tool cost, and one-time setup.
- Revenue uplift, major-incident avoidance, and customer-retention effects are excluded.

## Data

The companion Python model contains the complete scenario inputs and formulas.

```python
from pathlib import Path
import runpy

model_path = Path.cwd() / "docs/analysis/council-steward-roi-model-2026-07-24.py"
model = runpy.run_path(model_path)
scenario_results = model["scenario_results"]
weighted_expected = model["weighted_expected"]
scenario_results
```

## Results

```python
for row in scenario_results:
    print(
        row["scenario"],
        f'ROI={row["year_one_roi"]:.1%}',
        f'net=${row["year_one_net_benefit"]:,.0f}',
        f'break-even={row["break_even_benefit_hours_per_month"]:.2f} benefit hours/month',
    )
print("Weighted expected:", weighted_expected)
```

## Takeaways

```python
base = next(row for row in scenario_results if row["scenario"] == "Base")
print(f'Base year-one ROI: {base["year_one_roi"]:.0%}')
print(f'Base year-one net benefit: ${base["year_one_net_benefit"]:,.0f}')
print(f'Weighted expected ROI: {weighted_expected["expected_year_one_roi"]:.0%}')
print("Treat all results as provisional until the 30-day pilot measures actual time, findings, corrections, and false positives.")
```
